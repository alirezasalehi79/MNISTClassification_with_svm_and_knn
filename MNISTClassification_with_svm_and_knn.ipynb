```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

# **showing an image with matplotlib**

```python
data = datasets.load_digits()
plt.imshow(data.images[0],cmap='binary')
plt.show()
```

```python
images_and_labels = list(zip(data.images,data.target))
images_and_labels
```

```python
for (image,label) in images_and_labels[:5]:
  print(image,label)
```

```python
X = data.images.reshape((len(data.images),-1))
np.shape(X)
```

```python
y = data.target
print(len(y))
```

# **SVM algorithm**

```python
svm_classifier = SVC()
```

```python
svm_classifier = svm_classifier.fit(X,y)
```

```python
svm_classifier.predict([X[200],X[190]])
```

```python
u = data.images[190]
plt.imshow(u,cmap = plt.cm.gray_r)
plt.show()
```

# **Changing the numbers of some pixels**

```python
u[0,1] = 10
u[3,2] = 16
u[6,3] = 8
u[6,1] = 0
u
```

```python
plt.imshow(u,cmap = plt.cm.gray_r)
plt.show()
```

```python
u = u.reshape(-1)
svm_classifier.predict([u])
```

```python
svm_new = SVC()
svm_new = svm_new.fit(X[:1000],y[:1000])
```

```python
p = svm_new.predict(X[1000:])
```

```python
e = y[1000:]
```

# **compute accuracy,precision,recall and f1 score**

```python
print(metrics.classification_report(e,p))
```

```python
print(metrics.confusion_matrix(e,p))
```

# **KNN Classifier**

```python
data2 = datasets.load_digits()
np.set_printoptions(threshold=sys.maxsize)
```

# **flat samples**

```python
X = data2.images.reshape((1797,-1))
```

```python
y = data2.target
```

# **import algorithm**

```python
knn_clf = KNeighborsClassifier()
knn_clf = knn_clf.fit(X[:1000],y[:1000])
```

```python
p = knn_clf.predict(X[1000:])
e = y[1000:]
```

# **classification report and confusion matrix**

```python
print(metrics.classification_report(e,p))
```

```python
print(metrics.confusion_matrix(e,p))
```
